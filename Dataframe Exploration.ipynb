```python
solo_df = dataset[dataset['matchType'].str.contains('solo')]
solo_df['walkDistance'] = pd.cut(solo_df['walkDistance'], [0,500,1000,1500,2000,2500,3000,3500,4000], labels = ['0-500m', '500-1000m', '1000-1500m','1500-2000m','2000-2500m','2500-3000m','3000-3500m','3500-4000m'])
plt.figure(figsize=(40, 20))
sns.boxplot(x = 'walkDistance', y = 'winPlacePerc', hue = 'matchType', data=solo_df)
```

```python
solo_df = dataset[dataset['matchType'].str.contains('solo')]
solo_df['walkDistance'] = pd.cut(solo_df['walkDistance'], [0,500,1000,1500,2000,2500,3000,3500,4000], labels = ['0-500m', '500-1000m', '1000-1500m','1500-2000m','2000-2500m','2500-3000m','3000-3500m','3500-4000m'])
plt.figure(figsize=(40, 20))
sns.violinplot(x = 'walkDistance', y = 'winPlacePerc', data = solo_df)
plt.show()
plt.figure(figsize=(40, 20))
sns.boxplot(x = 'walkDistance', y = 'winPlacePerc', data = solo_df)
plt.show()
```

```python
tmp_df= dataset.copy()
tmp_df['kills'] = pd.cut(tmp_df['kills'], [-1,0,2,5,10,20,100], labels = ['0kills', '1-2kills', '3-5kills','6-10kills','11-20kills','21+kills'])
plt.figure(figsize=(15, 5))
sns.boxplot(x = 'kills', y = 'winPlacePerc', data=tmp_df)
```

```python
plt.figure(figsize=(40,20))
sns.countplot(dataset['kills'])
```

```python
plt.figure(figsize=(40, 20))
sns.jointplot(x = 'winPlacePerc', y = 'heals', data = dataset, height = 10, ratio = 3)
plt.ticklabel_format(style='plain',axis = 'both')
```

```python
tmp_df= dataset.copy()
plt.figure(figsize = (40,18))
sns.pointplot(x = 'heals', y = 'winPlacePerc', data=tmp_df, color = 'blue', alpha = 1.0)
sns.pointplot(x = 'boosts', y = 'winPlacePerc', data=tmp_df, color = 'lime', alpha = 1.0)
plt.xlabel('Number of Boosts/Heals Used', fontsize = 20, color = 'blue')
plt.ylabel('WinPlacePerc', fontsize = 20, color = 'blue')
plt.text(4, 0.9, 'Heals', color = 'lime', fontsize = 20)
plt.text(4, 0.6, 'Boosts', color = 'blue', fontsize = 20)
plt.title('Heals vs Boosts affecting winPlacePerc')
plt.grid()
```

```python
plt.figure(figsize = (40,18))
df_max = dataset.corr()
sns.set(font_scale =2)
sns.heatmap(df_max,annot = True, annot_kws ={"size":20}, linecolor = "white", linewidth = "2")
```

```python
plt.figure(figsize = (40,18))
cf_max = dataset.copy()
cf_max = cf_max.drop(columns = ['winPoints','killPoints','matchDuration', 'maxPlace'])
cf_max = cf_max.corr()
sns.set(font_scale =2)
sns.heatmap(cf_max,annot = True, annot_kws ={"size":20}, linecolor = "white", linewidth = "2")
```

```python
print('Average number of heals used is {: .1f}, 99% of player used {: .1f} heals or less, the max number of heals used was {: .1f}' .format(dataset['heals'].mean(), dataset['heals'].quantile(0.99), dataset['heals'].max()))
```
