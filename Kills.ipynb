```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_csv('/Users/User/Desktop/PUBG_match_data.csv')
```

```python
dataset[['kills', 'killStreaks']].describe()
```

```python
print('Average number of kills was {: .1f}, 99% of players had {: .1f} kills, the max number of kills was {: .1f}' .format(dataset['kills'].mean(), dataset['kills'].quantile(0.99), dataset['kills'].max()))

print('')

print('Average killstreak used was {: .1f}, 99% of players had a {: .1f} killstreak or less, the max killstreak was {: .1f}' .format(dataset['killStreaks'].mean(), dataset['killStreaks'].quantile(0.99), dataset['killStreaks'].max()))
```

```python
plt.figure(figsize=(40,20))
sns.histplot(data = dataset, x = 'kills', kde = True)
plt.ticklabel_format(style='plain', axis='y')
plt.show()
plt.figure(figsize=(40,20))
sns.countplot(data = dataset, x = 'killStreaks')
plt.ticklabel_format(style='plain', axis='y')
```

```python
plt.figure(figsize=(40,20))
sns.histplot(data = dataset, x = 'kills', kde = False, bins=24)
plt.ticklabel_format(style='plain', axis='y')
plt.show()
# plt.figure(figsize=(40,20))
# sns.countplot(data = dataset, x = 'killStreaks')
# plt.ticklabel_format(style='plain', axis='y')
```

```python
tmp_df= dataset.copy()
tmp_df['kills'] = pd.cut(tmp_df['kills'], [-1,0,2,5,10,75], labels = ['0kills', '1-2kills', '3-5kills','6-10kills','11+kills'])
plt.figure(figsize=(40, 20))
sns.violinplot(x = 'kills', y = 'winPlacePerc', data=tmp_df)
plt.show()
plt.figure(figsize=(40, 20))
sns.boxplot(x = 'kills', y = 'winPlacePerc', data = tmp_df)
plt.show()
```

```python
tmp_df= dataset.copy()
tmp_df['killStreaks'] = pd.cut(tmp_df['killStreaks'], [-1,0,2,5,21], labels = ['0 killstreak', '1-2 killsteak', '3-5 killstreak','6+ killstreak'])
plt.figure(figsize=(40, 20))
sns.violinplot(x = 'killStreaks', y = 'winPlacePerc', data=tmp_df)
plt.show()
plt.figure(figsize=(40, 20))
sns.boxplot(x = 'killStreaks', y = 'winPlacePerc', data = tmp_df)
plt.show()
```

```python
plt.figure(figsize=(40, 20))
sns.jointplot(x = 'winPlacePerc', y = 'kills', data = dataset, height = 10, ratio = 3)
plt.ticklabel_format(style='plain',axis = 'both')
plt.show()
plt.figure(figsize=(40, 20))
sns.jointplot(x = 'winPlacePerc', y = 'killStreaks', data = dataset, height = 10, ratio = 3)
plt.ticklabel_format(style='plain',axis = 'both')
plt.show()
```
