```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_csv('/Users/User/Desktop/PUBG_match_data.csv')
```

```python
dataset[['walkDistance', 'swimDistance', 'rideDistance']].describe()
```

```python
print('Average walking distance was {: .1f}, 99% of players had a walking distance of {: .1f} , the max walking distance {: .1f}' .format(dataset['walkDistance'].mean(), dataset['walkDistance'].quantile(0.99), dataset['walkDistance'].max()))

print('')

print('Average swimming distance was {: .1f}, 99% of players had a swimming distance of {: .1f} , the max swimming distance {: .1f}' .format(dataset['swimDistance'].mean(), dataset['swimDistance'].quantile(0.99), dataset['swimDistance'].max()))

print('')

print('Average vehicle distance was {: .1f}, 99% of players had a vehicle distance of {: .1f} , the max vehicle distance {: .1f}' .format(dataset['rideDistance'].mean(), dataset['rideDistance'].quantile(0.99), dataset['rideDistance'].max()))
```

```python
plt.figure(figsize=(40, 20))
sns.jointplot(x = 'winPlacePerc', y = 'walkDistance', data = dataset, height = 10, ratio = 3)
plt.ticklabel_format(style='plain',axis = 'both')
plt.show()
plt.figure(figsize=(40, 20))
sns.jointplot(x = 'winPlacePerc', y = 'swimDistance', data = dataset, height = 10, ratio = 3)
plt.ticklabel_format(style='plain',axis = 'both')
plt.show()
plt.figure(figsize=(40, 20))
sns.jointplot(x = 'winPlacePerc', y = 'rideDistance', data = dataset, height = 10, ratio = 3)
plt.ticklabel_format(style='plain',axis = 'both')
plt.show()
```
